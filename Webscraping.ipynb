```
import pandas as pd
import requests
import numpy as np
import sklearn 
import seaborn as sns
import matplotlib.pyplot as plt 
from bs4 import BeautifulSoup
import plotly.express as px
```

```
# send request

url ='https://www.booking.com/searchresults.pt-br.html?label=gen173nr-1DCAEoggI46AdILVgEaCCIAQGYAS24ARfIAQzYAQPoAQH4AQKIAgGoAgO4ApSE6PEFwAIB;sid=f13831b8dd2e86cde396fb0fc4c5e9b2;city=-649584&'
page = requests.get(url)
page.text
```

```
soup = BeautifulSoup(page.text,'html.parser')
```

```
# read hotel text descrisao 
booking_text = []
booking_text_elem = soup.find_all(class_="hotel_desc")
for item in booking_text_elem:
    booking_text.append(item.text)
booking_text
```

```
# read hotel name
hotel_name = []
hotel_name_elem = soup.find_all(class_="sr-hotel__name")
for item in hotel_name_elem:
    hotel_name.append(item.text)
hotel_name
```

```
#classification hotel

hotel_classification = []
hotel_classification_elem = soup.find_all(class_="bui-review-score__title")
for item in hotel_classification_elem:
    hotel_classification.append(item.text)
hotel_classification

```

```
# Avaliaçao_hotel
Avaliaçao_hotel = []
Avaliaçao_hotel_elem = soup.find_all(class_="bui-review-score__text")
for item in Avaliaçao_hotel_elem:
    Avaliaçao_hotel.append(item.text)
Avaliaçao_hotel

```

```
#localizaçao hotel

localizaçao_hotel = []
localizaçao_hotel_elem = soup.find_all(class_="review-score-badge")
for item in localizaçao_hotel_elem:
    localizaçao_hotel.append(item.text)
localizaçao_hotel
```

```
#creat dataframe
final_array = []
for text,hotel,classification,avaliaçao,localizaçao in zip(booking_text,hotel_name,hotel_classification,Avaliaçao_hotel,localizaçao_hotel):
  final_array.append({'Texte':text,    'Hotel':hotel,   'Classification':classification,   'Avaliaçao':avaliaçao,   'Localizaçao':localizaçao})
df = pd.DataFrame(final_array)

df.head(20)


```

```
fig = px.bar(df, x='Avaliaçao', y='Hotel')
fig.show()

fig = px.bar(df, x='Classification', y='Hotel')
fig.show()
```
